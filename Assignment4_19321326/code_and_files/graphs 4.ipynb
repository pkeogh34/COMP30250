```python
df_experiment3 = pd.read_csv('experiment3.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Getting unique matrix sizes
matrix_sizes = df_experiment3['Matrix Size'].unique()

# Define a color palette with NumPy
colors = plt.cm.viridis(np.linspace(0, 1, len(matrix_sizes)))

# Effects of Chunk Size with more colors
plt.figure(figsize=(14, 6))

for i, size in enumerate(matrix_sizes):
    subset = df_experiment3[df_experiment3['Matrix Size'] == size]
    avg_speedup = (subset['Speedup Static'] + subset['Speedup Dynamic'] + subset['Speedup Guided']) / 3
    plt.plot(subset['Chunk Size'], avg_speedup, marker='o', label=f'Matrix Size {size}', color=colors[i])

plt.title('Effects of Chunk Size on Average Speedup')
plt.xlabel('Chunk Size')
plt.ylabel('Average Speedup')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def plot_schedule_effects(matrix_size, df):
    """
    Plots the effects of different schedules on speedup for a given matrix size.
    
    Parameters:
    matrix_size: int
        The matrix size for which to plot the schedule effects.
    df: pandas.DataFrame
        The DataFrame containing the data.
    """
    # Filter the DataFrame for the specified matrix size
    subset = df[df['Matrix Size'] == matrix_size]
    
    # Plotting
    plt.figure(figsize=(14, 6))
    plt.plot(subset['Chunk Size'], subset['Speedup Static'], marker='o', label='Static', color='b')
    plt.plot(subset['Chunk Size'], subset['Speedup Dynamic'], marker='^', label='Dynamic', color='g')
    plt.plot(subset['Chunk Size'], subset['Speedup Guided'], marker='s', label='Guided', color='r')
    plt.title(f'Effects of Schedule on Speedup (Matrix Size {matrix_size})')
    plt.xlabel('Chunk Size')
    plt.ylabel('Speedup')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
# Example usage
plot_schedule_effects(8, df_experiment3)  # Replace 32 with any matrix size you want to analyze
```
